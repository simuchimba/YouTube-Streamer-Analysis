```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
```

```python
df = pd.read_csv('youtubers_df.csv')
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
if 'Categories' in df.columns and missing_values['Categories'] > 0:

  df.drop(['Categories'], axis=1, inplace=True)

  df = df.dropna(axis=1)

  print(df.isnull().sum())
```

```python
statistics = df.describe()
print("statistics ", statistics)
```

```python
min_values = statistics.loc['min']
max_values = statistics.loc['max']
quartiles = statistics.loc[['25%', '50%', '75%']]

print("Below are the minimum values:")
print(min_values)

print("\nBelow are the maximum values:")
print(max_values)

print("\nBelow are the quartile values:")
print(quartiles)
```

```python
numeric_columns = df.select_dtypes(include=['number'])

Q1 = numeric_columns.quantile(0.25)
Q3 =  numeric_columns.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = ((numeric_columns < lower_bound )| (numeric_columns > upper_bound ))
outliers
```

```python
plt.scatter(numeric_columns.index, numeric_columns['Suscribers'], c=outliers['Suscribers'], cmap='viridis')
plt.xlabel('Index')
plt.ylabel('Subscribers')
plt.title('Outliers Detection using IQR')
plt.show()
```

```python
plt.scatter(numeric_columns.index, numeric_columns['Comments'], c=outliers['Comments'], cmap='viridis')
plt.xlabel('Index')
plt.ylabel('Subscribers')
plt.title('Outliers Detection using IQR')
plt.legend()
plt.show()
```

```python
plt.scatter(numeric_columns.index, numeric_columns['Likes'], c=outliers['Likes'], cmap='viridis')
plt.xlabel('Index')
plt.ylabel('Subscribers')
plt.title('Outliers Detection using IQR')
plt.legend()
plt.show()
```

```python
plt.scatter(numeric_columns.index, numeric_columns['Visits'], c=outliers['Visits'], cmap='viridis')
plt.xlabel('Index')
plt.ylabel('Subscribers')
plt.title('Outliers Detection using IQR')
plt.legend()
plt.show()
```

```python
df_no_outliers = df[~outliers]
```

```python
dT = pd.read_csv('youtubers_df.csv')

numeric_columns = ['Suscribers', 'Visits', 'Likes', 'Comments']

category_trends = dT.groupby('Categories')[numeric_columns].mean()
category_trends = category_trends.sort_values(by='Suscribers', ascending=False)

plt.figure(figsize=(10,6))
category_trends.plot(kind='bar', stacked=True)
plt.title('Average Matrics by Categories')
plt.ylabel('Average Value')
plt.xlabel('Category')
plt.show()
```

```python
correlation_matrix = df[numeric_columns].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Suscribers' , y='Likes' , data=df)
plt.title('Suscribers vs Likes')
plt.xlabel('Suscribers')
plt.ylabel('Likes')
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Suscribers' , y='Comments', data=df)
plt.title('Suscribers vs Comments')
plt.xlabel('Suscribers')
plt.ylabel('Comments')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

country_distribution = dT['Country'].value_counts()

plt.figure(figsize=(12, 8))
country_distribution.plot(kind='bar')
plt.title('Distribution of Streams by Country')
plt.xlabel('Country')
plt.ylabel('Number of Streamers')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.countplot(x='Country', hue='Categories', data=dT)
plt.title('Regional preferences for content Categories')
plt.xlabel('Country')
plt.ylabel('Number of streams')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

avg_subscribers = dT['Suscribers'].mean()
avg_visits = dT['Visits'].mean()
avg_likes = dT['Likes'].mean()
avg_comments = dT['Comments'].mean()

metrics = ['Suscribers', 'Visits', 'Likes', 'Comments']
average_values = [avg_subscribers, avg_visits, avg_likes, avg_comments]

plt.figure(figsize=(10, 6))
plt.bar(metrics, average_values, color=['blue', 'green', 'orange', 'red'])
plt.title('Average Performance Metrics')
plt.ylabel('Average Values')
plt.show()
```

```python
category_distribution = dT['Categories'].value_counts()

plt.figure(figsize=(10, 6))
category_distribution.plot(kind='bar', color='green')
plt.title('Distribution of Content Categories')
plt.xlabel('Content Categories')
plt.ylabel('Number of Streamers')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
metric_columns = ['Suscribers', 'Visits', 'Likes', 'Comments']

average_metrics_by_category = dT.groupby('Suscribers')[metric_columns].mean()

plt.figure(figsize=(20, 12))
average_metrics_by_category.plot(kind='bar', colormap='viridis')
plt.title('Average Performance Metrics')
plt.xlabel('Content Categories')
plt.ylabel('Average Value')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Metrics')
plt.show()
```

```python
metric_columns = ['Suscribers', 'Visits', 'Likes', 'Comments']

average_metrics = dT[metric_columns].mean()

above_average_streamers = dT[(dT['Suscribers'] > average_metrics['Suscribers'])
                              & (dT['Visits'] > average_metrics['Visits'])
                              &(dT['Likes'] > average_metrics['Likes'])
                              &(dT['Comments'] > average_metrics['Comments'])]

top_performing_creators = above_average_streamers[['Username', 'Suscribers', 'Visits', 'Likes', 'Comments']]
top_performing_creators = top_performing_creators.sort_values(by='Suscribers', ascending=False)

print("Top-performing content creators:")
print(top_performing_creators)
```
